```python
import pandas as pd
from textblob import TextBlob as tb
from string import punctuation
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
df = pd.read_pickle('C:/Users/Samarth/Desktop/Mtech AI/NLP/Data/first_pickle.zip')
```

```python
desc = df['Consumer complaint narrative'].str.lower()
```

```python
desc = desc.dropna()
```

```python
desc = desc[:10000]
```

```python
stopword = set(stopwords.words('english')+list(punctuation)+['xxx','xxxx'])
```

```python
desc = pd.DataFrame(desc)
```

```python
blob = tb(desc['Consumer complaint narrative'].iloc[0])
```

```python
blob.sentiment[0]
```

```python
sentiment = []
for x in range(len(desc['Consumer complaint narrative'])):
    blob= tb(desc['Consumer complaint narrative'].iloc[x])
    sentiment.append(blob.sentiment[0])
```

```python
desc['Sentiment'] = sentiment
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count = CountVectorizer(stop_words='english')
```

```python
x = desc['Consumer complaint narrative']
x = count.fit_transform(x)
```

```python
x.shape
```

```python
y = desc['Sentiment']
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.metrics import r2_score
```

```python
tree = DecisionTreeRegressor()
```

```python
params = {'max_depth':np.arange(1,10,1), 'min_samples_leaf':np.arange(1,10,1)}
```

```python
grid = GridSearchCV(tree,param_grid=params,cv = 10,verbose=3,scoring='r2')
```

```python
grid.fit(x,y)
```

```python
grid.best_score_
```
