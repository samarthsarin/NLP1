```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from string import punctuation
from collections import Counter
from nltk.corpus import reuters
import operator
import pandas as pd

import spacy
nlp = spacy.load('en_core_web_sm')

stop_words = list(set(stopwords.words('english')))+list(punctuation)
lem = WordNetLemmatizer()

def scoreing(file):
    with open(file,'r') as f:
        text = f.read()
    clean_doc = cleaning(text)

    clean_doc_1 = ' '.join(clean_doc)

    tokens = []
    doc = nlp(clean_doc_1)
    for token in doc:
        if not token.is_stop and not token.is_digit and not token.is_punct:
            tokens.append(token.text)

    tokens = ' '.join(tokens)

    green_words = []
    for i in fields:
        words  = reuters.words(i)
        for w in words:
            green_words.append(w.lower())

    clean_doc = cleaning(tokens)

    count = dict(Counter(clean_doc))

    count = dict(sorted(count.items(),key=operator.itemgetter(1),reverse=True))

    words_doc = [list(count.keys())[i] for i in range(len(count))]
    count_doc = [list(count.values())[i] for i in range(len(count))]

    useful_words = []
    useful_count = []
    for i in range(len(words_doc)):
        if words_doc[i] in green_words:
            useful_words.append(words_doc[i])
            useful_count.append(count_doc[i])

    total_words = len(words_doc)
    total_useful = len(useful_count)
    score = total_useful/total_words
    score = int(score*1000)

    return score
```

```python
def cleaning(lines):
    lines = lines.lower()
    words = word_tokenize(lines)
    words = [w for w in words if len(w)>3]
    words_stop = [w for w in words if w not in stop_words]
    words_lem =[lem.lemmatize(w,'v') for w in words]
    return words_lem
```
