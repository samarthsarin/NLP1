```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from string import punctuation
from nltk.corpus import stopwords
from textblob import TextBlob as tb
import pandas as pd
```

```python
stop_words = list(set(stopwords.words('english')))+list(punctuation)
```

```python
lem = WordNetLemmatizer()
def cleaning(text):
    words = text.lower()
    words = word_tokenize(words)
    words = [t for t in words if t not in stop_words]
    lemma = [lem.lemmatize(w,'v') for w in words]
    words = ' '.join(lemma)
    return words
```

```python
df = pd.read_csv('C:/Users/Samarth/Desktop/Mtech AI/NLP/Data/haptik_open_datasets-master/domain_classification/train_data.csv')
```

```python
df['Label'] = ' '
```

```python
df = df.drop(['food','recharge','support','reminders','travel','nearby','movies','casual','other'],1)
```

```python
df['message'] = df['message'].apply(cleaning)
```

```python
for i in range(len(df['message'])):
    blob = tb(df['message'].iloc[i])
    df['Label'].iloc[i] = blob.sentiment[0]
```

```python
df.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
senti = SentimentIntensityAnalyzer()
```

```python
df.head()
```

```python
df['Vader Score'] = ' '
```

```python
vader = []
for i in range(len(df)):
    vader.append(senti.polarity_scores(df['message'].iloc[i]))
df['Vader Score'] = vader
```

```python
df.head()
```

```python
df['Vader Compund'] = ' '
```

```python
for i in range(df.shape[0]):
    df['Vader Compund'].iloc[i] = df['Vader Score'].iloc[i]['compound']
```
